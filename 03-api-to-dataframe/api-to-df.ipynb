```python
import pandas as pd 
import requests
```

```python
response = requests.get('https://pharma-server-07.vercel.app/api/medicine?page=1')
```

```python
page = response.json()['data']['totalPages']
page
```

```python
medicines = pd.DataFrame(response.json()['data']['medicines'])
medicines.head(2)
```

```python
# Normalize manufacturer fields
manufacturer_df = pd.json_normalize(medicines['manufacturer'])
```

```python
# Rename manufacturer columns
manufacturer_df.columns = ['manufacturer_' + col for col in manufacturer_df.columns]
```

```python
pd.concat([medicines[['_id', 'name', 'category','price']], manufacturer_df[['manufacturer_name','manufacturer_contact']]], axis=1)
```

```python
df = pd.DataFrame()
all_data = []

for i in range(1, page+1):
    response = requests.get('https://pharma-server-07.vercel.app/api/medicine?page={}'.format(i))
    temp_df = pd.DataFrame(response.json()['data']['medicines'])
    all_data.append(temp_df)

df = pd.concat(all_data, ignore_index=True)
# df.head(2)

all_manufacturer_df = pd.json_normalize(df['manufacturer'])
all_manufacturer_df.columns = ['manufacturer_' + col for col in all_manufacturer_df.columns]
df = pd.concat([df[['_id', 'name', 'category','price']], all_manufacturer_df[['manufacturer_name','manufacturer_contact']]], axis=1)
df.shape
```
