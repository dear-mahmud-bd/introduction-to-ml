```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

# 1. Categorical Data

## a. Countplot

```python
sns.countplot(x='Pclass', data=df)
# df['Survived'].value_counts().plot(kind='bar')
```

## b. PieChart

```python
df['Pclass'].value_counts().plot(kind='pie',autopct='%.2f')
```

# 2. Numerical Data

## a. Histogram

```python
import matplotlib.pyplot as plt
counts, bins, patches = plt.hist(df['Age'],bins=8, edgecolor='black')
plt.xticks(bins.round(1))
plt.show()
```

## b. Distplot

```python
# stat= "count", "percent", "probability"
sns.displot(df['Age'], kde=True, bins=8, stat="density") # (kernel density estimation) kde=True` adds the smooth curve
plt.show()
```

## c. Boxplot

```python
sns.boxplot(df['Age'])
plt.show()
```
